```python
columns = ['diagnosis',
        'radius_mean', 
        'texture_mean', 
        'perimeter_mean', 
        'area_mean', 
        'smoothness_mean', 
        'compactness_mean', 
        'concavity_mean',
        'concave points_mean', 
        'symmetry_mean', 
        'fractal_dimension_mean']
```

```python
sns.pairplot(data_set[columns],hue = 'diagnosis')
```

```python
x = data_set[['radius_mean','area_mean','concavity_mean','concave points_mean','perimeter_mean']].values
```

```python
y = data_set['diagnosis'].values
```

```python
#Veri setimiz üzerinde bölme işlemlerini yapalım
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
#Özellik ölçeklendirme
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

X_train = ss.fit_transform(x_train)
X_test = ss.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = 'gini',random_state=29)
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
from tabulate import tabulate
report = classification_report(y_test,y_pred,target_names=['B', 'M'])
table = tabulate([line.split() for line in report.split('\n') if line], headers=['', 'precision', 'recall', 'f1-score', 'support'], tablefmt='pretty')
print(table)
```
